# Customer

```python
import pandas as pd

df = pd.read_csv("../New folder/Project_Assignment/Cust_Purch_FakeData.csv")
df1 = pd.read_csv('../numpy_practice/jupyter/pandas_practice/Movies_dataset.csv')
df2 = pd.read_csv('../numpy_practice/jupyter/pandas_practice/top_250_imdb_dataset.csv')
df3 = pd.read_csv('../numpy_practice/jupyter/pandas_practice/Students Social Media Addiction.csv')
```

```python
# 1Ô∏è‚É£ Age 25‚Äì35 AND purchase 80‚Äì100

(
    df[(df['age'].between(25, 35)) &
        (df['price(CAD)'].between(80, 100))]
        .reset_index( drop = True)
)
```

```python
# 2Ô∏è‚É£ High spenders from Male

high_spender_male = (
    df[(df['price(CAD)'] > 90) &
        (df['gender'] == 'Male')]
)
high_spender_male.reset_index( drop = True)
```

```python
# 3Ô∏è‚É£ Customers who purchased in 2020 AND spent > 95

df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year

vip_customers = (
    df[(df['year'] == 2020) &
        (df['price(CAD)'] > 95)]
)
vip_customers.reset_index( drop = True)
```

# Movies

```python
# 4Ô∏è‚É£ Duration 100‚Äì180 AND year > 2015

(
    df2[(df2['Duration_Minutes'].between(100, 180)) &
        (df2['Year'] > 2015)]
        .reset_index(drop = True)
)
```

```python
# 5Ô∏è‚É£ Votes > 100 AND popularity > 120

(
    df1[(df1['vote_count'] > 100) &
        (df1['popularity'] > 120)]
        .reset_index()
)
```

```python
# 6Ô∏è‚É£ Movies between 2000‚Äì2010 sorted by votes

#  first remove the null values of release_date.
df1.dropna(subset = ['release_date'], inplace = True)

# Now convert obj data type into datetime data type.
df1['release_date'] = pd.to_datetime(df1['release_date'])
df1['release_year'] = df1['release_date'].dt.year

(
    df1[df1['release_year'].between(2000, 2010)]
       .sort_values(by = 'vote_count', ascending = False)
       .reset_index(drop = True)
)
```

# Students

```python
# 7Ô∏è‚É£ Addiction ‚â• 8 AND sleep < 6 AND usage > 6

Addicted_students = (
    df3[(df3['Addicted_Score'] >= 8) &
        (df3['Sleep_Hours_Per_Night'] < 6) &
        (df3['Avg_Daily_Usage_Hours'] > 6)]
)
Addicted_students.reset_index( drop = True)
```

```python
# 8Ô∏è‚É£ Female students with addiction ‚â• 7

female_students = (
    df3[(df3['Gender'] == 'Female') &
        (df3['Addicted_Score'] >= 7)]
)
female_students.reset_index(drop = True)
```

```python
# 9Ô∏è‚É£ Bangladesh students with usage > 5

bangladeshi_students = (
    df3[(df3['Country'] == 'Bangladesh') &
        (df3['Avg_Daily_Usage_Hours'] > 5)]
)
bangladeshi_students.reset_index( drop = True)
```

# üëâ ‚ÄúTop 5 most voted movies  between 2000‚Äì2010‚Äù

```python
(
    df1[df1['release_year'].between(2000, 2010)]
       .groupby('release_year')['vote_count']
       .max()
       .sort_values( ascending = False) 
       .head(5)
       .reset_index(name = 'top_5_voted_movies')
       
)
```

# üëâ ‚ÄúTop 5 most voted movies per year between 2000‚Äì2010‚Äù

```python
(
    df1[df1['release_year'].between(2000, 2010)]
       .sort_values(['release_year','vote_count'] ,ascending = [ True , False ])
       .groupby('release_year')
       .head(5)       
)
```
