# ü•â DAY 3 ‚Äî Sorting & Value Counts

```python
import pandas as pd

df = pd.read_csv("../New folder/Project_Assignment/Cust_Purch_FakeData.csv")
df1 = pd.read_csv('../numpy_practice/jupyter/pandas_practice/Movies_dataset.csv')
df2 = pd.read_csv('../numpy_practice/jupyter/pandas_practice/top_250_imdb_dataset.csv')
df3 = pd.read_csv('../numpy_practice/jupyter/pandas_practice/Students Social Media Addiction.csv')
```

# Customer

```python
# 1Ô∏è‚É£ Top 10 highest purchase

(
    df['price(CAD)']
        .sort_values(ascending = False)
        .head(10)
        .reset_index(name = 'top_10_purchase')
)
```

```python
# 2Ô∏è‚É£ Lowest 5 purchases

(
    df['price(CAD)']
        .sort_values(ascending = True)
        .head(5)
        .reset_index(name = 'Lowest_5_purchases')
)
```

```python
# 3Ô∏è‚É£ Gender count

(
    df['gender']
        .value_counts()
        .reset_index(name = 'total')
)
```

```python
# 4Ô∏è‚É£ Country count

( 
    df3['Country']
        .value_counts()
        .reset_index(name = 'total')
)
```

# Movies

```python
# 5Ô∏è‚É£ Top 10 longest movies

(
    df2['Duration_Minutes']
         .sort_values( ascending = False)
         .head(10)
         .reset_index(name = 'top_10_longest_movie')
)
```

```python
# 6Ô∏è‚É£ Top 5 most voted movies

(
    df1['vote_count']
        .sort_values(ascending = False)
        .head(5)
        .reset_index(name = 'top_5_voted_movies')
)
```

```python
# 7Ô∏è‚É£ Movies per year (value_counts)

(
   df2['Year']
       .value_counts()
       .reset_index(name = 'total_movies_per_year')
)
```

# Students

```python
# 8Ô∏è‚É£ Platform usage count

(
    df3['Most_Used_Platform']
        .value_counts()
        .reset_index(name = 'total')
)
```

```python
# 9Ô∏è‚É£ Top 10 addicted students

(
    df3[['Student_ID' ,'Addicted_Score']]
         .sort_values(by = 'Addicted_Score', ascending = False)
         .head(10)
)
```

```python
# üîü Top 10 daily usage students

(
    df3[['Student_ID' , 'Avg_Daily_Usage_Hours']]
         .sort_values(by = 'Avg_Daily_Usage_Hours', ascending = False )
         .head(10)
)
```
