# Dataset: Movies

```python
# 8Ô∏è‚É£ Total number of movies

total = df1['id'].count()
print(f'Total movies: {total}')
```

```python
# 9Ô∏è‚É£ Unique genres count

unique_genre = df1['genre'].unique().shape[0]
print(f'Total unique genre: {unique_genre}')
```

```python
# üîü Oldest movie year

oldest =(
    df1['release_date']
        .sort_values()
        .head(1)
        .reset_index(name = 'oldest_movie')
)
oldest
```

```python
# 1Ô∏è‚É£1Ô∏è‚É£ Newest movie year

newest = (
    df1['release_date']
        .sort_values(ascending = False)
        .head(1)
        .reset_index(name = 'newest_movie')
)
newest
```

# Dataset: Students

```python
# 1Ô∏è‚É£2Ô∏è‚É£ Total number of students

total = df2.shape[0]
print('Total student :',total)
```

```python
# 1Ô∏è‚É£3Ô∏è‚É£ Unique countries

countries = df2['Country'].unique()
print('Total unique countries: ',countries)
```

```python
# 1Ô∏è‚É£4Ô∏è‚É£ Average addiction score

avg = df2['Addicted_Score'].mean().round(2)
print('Average addicted score: ', avg)
```

```python
# 1Ô∏è‚É£5Ô∏è‚É£ Average sleep hours

avg_sleep = df2['Sleep_Hours_Per_Night'].mean().round(2)
print(f'Average Sleep hours: {avg_sleep}')
```
