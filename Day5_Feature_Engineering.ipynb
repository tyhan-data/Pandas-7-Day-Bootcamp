# Customer

```python
import pandas as pd

df = pd.read_csv("../New folder/Project_Assignment/Cust_Purch_FakeData.csv")
df1 = pd.read_csv('../numpy_practice/jupyter/pandas_practice/Movies_dataset.csv')
df2 = pd.read_csv('../numpy_practice/jupyter/pandas_practice/top_250_imdb_dataset.csv')
df3 = pd.read_csv('../numpy_practice/jupyter/pandas_practice/Students Social Media Addiction.csv')
```

```python
# 1Ô∏è‚É£ Convert purchase_date to datetime

df['date'] = pd.to_datetime(df['date'])
df['date'].dtypes
```

```python
# 2Ô∏è‚É£ Extract purchase year

df['year'] = df['date'].dt.year
df.head(1)
```

```python
# 3Ô∏è‚É£ Count purchases per year

(
    df
      .groupby('year')
      .size()
      .reset_index( name = 'count_purchases')
      .sort_values( by = 'count_purchases', ascending = False)   
)
```

```python
#4Ô∏è‚É£ Find purchases made in 2019

(
    df[df['year'] == 2019]
        [['first','last','price(CAD)','year']]
        .sort_values(by = 'price(CAD)', ascending = False)
        .reset_index()
)
```

```python
# 5Ô∏è‚É£ Create column:
# High_Spender ‚Üí purchase_amount > 80
# Low_Spender ‚Üí purchase_amount < 80


df['user_type'] = df['price(CAD)'].apply(
        lambda x : 'High_Spender' if x > 80 else 'Low_Spender' )

df.head()
```

# Movies

```python
# 6Ô∏è‚É£ Create column:
# Is_Long ‚Üí duration > 150

df2['Is_Long'] = df2['Duration_Minutes'] > 150
df2.head()
```

```python
# 7Ô∏è‚É£ Extract decade from year

df2['Decade'] = (df2['Year'] // 10) *10
df2.head()
```

```python
# 8Ô∏è‚É£ Count movies per decade

(
    df2
       .groupby('Decade')
        .size()
       .reset_index(name = 'total')
)
```

# Students

```python
# 9Ô∏è‚É£ Create Addiction_Level column:

# 0‚Äì4 ‚Üí Low

# 5‚Äì7 ‚Üí Medium

# 8‚Äì10 ‚Üí High



df3['Addiction_Level'] = df3['Addicted_Score'].apply(
    lambda a : 'Low' if 0 <= a <= 4  else 'Medium' if 5 <= a <= 7 else 'High'
)
df3.head()
```

```python
# üîü Count students per Addiction_Level

(
    df3
       .groupby('Addiction_Level')
        .size()
        .reset_index( name = 'Total_Student')
)
```
