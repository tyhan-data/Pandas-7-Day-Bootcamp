```python
import pandas as pd

df = pd.read_csv("../New folder/Project_Assignment/Cust_Purch_FakeData.csv")
df1 = pd.read_csv('../numpy_practice/jupyter/pandas_practice/Movies_dataset.csv')
df2 = pd.read_csv('../numpy_practice/jupyter/pandas_practice/top_250_imdb_dataset.csv')
df3 = pd.read_csv('../numpy_practice/jupyter/pandas_practice/Students Social Media Addiction.csv')
```

```python
# 1Ô∏è‚É£ Total purchase per profession

(
    df
        .groupby('profession')['price(CAD)']
        .sum()
        .reset_index(name = 'total')
        .sort_values(by = 'total', ascending = False)
       
)
```

```python
# 2Ô∏è‚É£ Average purchase per gender

(
    df
        .groupby('gender')['price(CAD)']
        .mean()
        .reset_index( name = 'avg_purchase')
        .sort_values(by = 'avg_purchase' , ascending = False)
)
```

```python
# 3Ô∏è‚É£ Average age per gender

(
   df
      .groupby('gender')['age']
      .mean()
      .reset_index(name = 'avg_age')
      .sort_values(by = 'avg_age', ascending = False)
)
```

```python
# 4Ô∏è‚É£ Company with highest total revenue

(
    df
      .groupby('company')['price(CAD)']
      .sum()
      .reset_index( name = 'highest_total_revenue')
      .sort_values(by = 'highest_total_revenue', ascending = False)
      .head(1)
)
```

```python
# 5Ô∏è‚É£ Revenue trend per weekday

(
    df
      .groupby('weekday')['price(CAD)']
      .sum()
      .reset_index(name = 'revenue_per_weekday')
      .sort_values( by ='revenue_per_weekday', ascending = False)
)
```

```python
# 6Ô∏è‚É£ High spender percentage per profession


(
    df
      .assign(high_spender = df['price(CAD)'] > 50)
      .groupby('profession')['high_spender']
      .mean()
      .sort_values(ascending=False) * 100
     
)
```

# üé¨ Movies Dataset Analysis

```python
# 7Ô∏è‚É£ Total movies

print(f'Total movies :{df1.shape[0]}')
```

```python
# 8Ô∏è‚É£ Movies per year

(
    df1
       .groupby('release_date').size()
       .reset_index( name = 'movies_per_year')
       .sort_values(by = 'movies_per_year', ascending = False)
)
```

```python
# 9Ô∏è‚É£ Average duration per year

(
    df2
       .groupby('Year')['Duration_Minutes']
       .mean()
       .reset_index(name = 'avg_duration')
       .sort_values( by = 'avg_duration' , ascending = False)
       .round(2)
)
```

```python
# üîü Most voted movies per language

(
    df1
       .groupby('original_language')['vote_count']
       .max()
       .reset_index(name = 'most_vote')
       .sort_values(by = 'most_vote', ascending = False)
)
```

```python
# 1Ô∏è‚É£1Ô∏è‚É£ Duration trend over time

duration_trend = (
    df2
       .groupby('Year')['Duration_Minutes']
       .mean()
       .reset_index()
)

import matplotlib.pyplot as plt

plt.figure()
plt.plot(duration_trend['Year'], duration_trend['Duration_Minutes'])
plt.xlabel('Year')
plt.ylabel('Average Duration')
plt.title('Movie Duration Trend Over Time')
plt.show()
```
